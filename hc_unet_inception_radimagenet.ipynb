```python
# Single-cell end-to-end Jupyter pipeline (no installs) - V6 (RadImageNet Backbone)

# ─── Imports & Configuration ──────────────────────────────────────────────────
import os, glob, cv2, numpy as np, pandas as pd
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
import copy

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import timm # Ensure timm is available
import torch.nn.functional as F
import torch.optim as optim
from torch.optim import lr_scheduler

# ─── UPDATE THESE PATHS TO YOUR SETUP ─────────────────────────────────────────
TRAIN_DIR  = "/home/sjim/Courses/Intro to ML/1327317/training_set"
TEST_DIR   = "/home/sjim/Courses/Intro to ML/1327317/test_set"
TRAIN_CSV  = "/home/sjim/Courses/Intro to ML/1327317/training_set_pixel_size_and_HC.csv"
TEST_CSV   = "/home/sjim/Courses/Intro to ML/1327317/test_set_pixel_size.csv"

# --- NEW: Path to your downloaded RadImageNet InceptionV3 weights ---
# <<<--- UPDATE THIS PATH --->>>
RADIMAGENET_INCEPTIONV3_PATH = "/home/sjim/Code/Pretrained_Rad_Models/RadImageNet_pytorch/InceptionV3.pt"

# --- Configuration ---
IMG_SIZE      = 256
BATCH_SIZE    = 8
EPOCHS        = 50
LR            = 1e-4
DEVICE        = torch.device("cuda" if torch.cuda.is_available() else "cpu")
DICE_WEIGHT   = 5.0
NUM_PLOT_EXAMPLES = 5

# --- Normalization constants (Assuming RadImageNet uses similar stats to ImageNet/general) ---
# If RadImageNet paper specifies different stats, use those.
NORM_MEAN = [0.5, 0.5, 0.5]
NORM_STD = [0.5, 0.5, 0.5]

print(f"Device: {DEVICE}, resizing to {IMG_SIZE}×{IMG_SIZE}")
print(f"Using combined loss: BCE + {DICE_WEIGHT}*DiceLoss")
print(f"Normalization: Mean={NORM_MEAN}, Std={NORM_STD}")
print(f"Epochs: {EPOCHS}, Initial LR: {LR}")
print(f"Attempting to load RadImageNet InceptionV3 weights from: {RADIMAGENET_INCEPTIONV3_PATH}")


# ─── CSV Loader (unchanged) ───────────────────────────────────────────────────
def load_csv(csv_path, want_hc=True):
    # (Code is unchanged, omitted for brevity)
    if not os.path.isfile(csv_path): raise FileNotFoundError(f"CSV file not found: {csv_path}")
    df = pd.read_csv(csv_path, index_col=0); df.index = df.index.astype(str)
    pix_cols = [c for c in df.columns if "pixel" in c.lower()]; assert pix_cols, f"No pixel-size column in {csv_path}"; px_col = pix_cols[0]
    hc_col = None
    if want_hc:
        hc_cols = [c for c in df.columns if ("hc" in c.lower() or "circumference" in c.lower()) and c != px_col]
        if not hc_cols: print(f"Warning: No head-circumference column found in {csv_path}. MAE cannot be calculated.")
        else: hc_col = hc_cols[0]
    return df, px_col, hc_col

# ─── Dataset (unchanged) ──────────────────────────────────────────────────────
class HCDataset(Dataset):
    # (Code is unchanged, omitted for brevity)
    def __init__(self, paths, df, px_col, hc_col=None, mode="train", img_size=256, norm_mean=None, norm_std=None):
        self.paths, self.df, self.px_col, self.hc_col, self.mode = paths, df, px_col, hc_col, mode; self.img_size = img_size; self.norm_mean = norm_mean if norm_mean else [0.5, 0.5, 0.5]; self.norm_std = norm_std if norm_std else [0.5, 0.5, 0.5]
        if self.mode == "train": self.img_augmentations = transforms.Compose([transforms.RandomHorizontalFlip(p=0.5), transforms.RandomRotation(degrees=10)]); # print("Applying image augmentations for training set.") # Less verbose
        else: self.img_augmentations = None
        self.img_base_transforms = transforms.Compose([transforms.Resize((self.img_size, self.img_size), interpolation=transforms.InterpolationMode.BILINEAR), transforms.ToTensor(), transforms.Lambda(lambda x: x.repeat(3, 1, 1)), transforms.Normalize(mean=self.norm_mean, std=self.norm_std)])
        if self.mode == "train": self.mask_augmentations = transforms.Compose([transforms.RandomHorizontalFlip(p=0.5), transforms.RandomRotation(degrees=10)]); # print("Applying mask augmentations for training set.") # Less verbose
        else: self.mask_augmentations = None
        self.mask_base_transforms = transforms.Compose([transforms.Resize((self.img_size, self.img_size), interpolation=transforms.InterpolationMode.NEAREST), transforms.ToTensor()])
    def __len__(self): return len(self.paths)
    def __getitem__(self, i):
        img_p = self.paths[i]; fn = os.path.basename(img_p)
        if fn not in self.df.index: print(f"Warning: Filename {fn} not found in DataFrame index. Skipping."); return None
        try: img = Image.open(img_p).convert("L")
        except Exception as e: print(f"Error loading image {img_p}: {e}"); return None
        mask = None
        if self.mode in ("train", "val"):
            m_p = img_p.replace("_HC.png", "_HC_Annotation.png")
            if os.path.exists(m_p):
                try: mask = Image.open(m_p).convert("L")
                except Exception as e: print(f"Error loading mask {m_p}: {e}. Using empty mask placeholder."); mask = Image.fromarray(np.zeros((self.img_size, self.img_size), dtype=np.uint8))
            else: mask = Image.fromarray(np.zeros((self.img_size, self.img_size), dtype=np.uint8))
        if self.mode == "train":
            state = torch.get_rng_state(); img = self.img_augmentations(img)
            if mask is not None: torch.set_rng_state(state); mask = self.mask_augmentations(mask)
        img_tensor = self.img_base_transforms(img)
        if mask is not None: mask_tensor = self.mask_base_transforms(mask); mask_tensor = (mask_tensor > 0.5).float()
        else: mask_tensor = torch.zeros(1, self.img_size, self.img_size, dtype=torch.float)
        px_sz = float(self.df.loc[fn, self.px_col]) if self.px_col else 0.0
        hc_gt = float(self.df.loc[fn, self.hc_col]) if self.hc_col and fn in self.df.index and self.hc_col in self.df.columns else -1.0
        return img_tensor, mask_tensor, fn, px_sz, hc_gt

# --- Custom collate function (unchanged) ---
def collate_fn_skip_none(batch):
    # (Code is unchanged, omitted for brevity)
    batch = list(filter(lambda x: x is not None, batch));
    if not batch: return None
    return torch.utils.data.dataloader.default_collate(batch)

# ─── Model Definition (MODIFIED to load RadImageNet weights) ────────────────
class DecoderBlock(nn.Module):
    # (Code is unchanged, omitted for brevity)
    def __init__(self, in_ch, out_ch): super().__init__(); self.up = nn.Upsample(scale_factor=2, mode="bilinear", align_corners=False); self.conv = nn.Sequential(nn.Conv2d(in_ch, out_ch, 3, padding=1, bias=False), nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True))
    def forward(self, x): return self.conv(self.up(x))

class UNetInceptionV3(nn.Module):
    def __init__(self, radimagenet_weights_path=None): # Added path argument
        super().__init__()

        # 1. Create InceptionV3 model structure without default weights, features only
        self.encoder = timm.create_model(
            "inception_v3",
            pretrained=False, # Important: Don't load ImageNet weights here
            features_only=True,
            in_chans=3
        )

        # 2. Load RadImageNet weights if path is provided
        if radimagenet_weights_path and os.path.exists(radimagenet_weights_path):
            try:
                # Load the state dict from the .pt file
                # Use map_location to load onto the correct device (e.g., CPU initially)
                rad_weights = torch.load(radimagenet_weights_path, map_location=torch.device('cpu'))

                # It's common for saved models to have an outer key like 'model' or 'state_dict'
                if isinstance(rad_weights, dict) and 'model' in rad_weights:
                    rad_weights = rad_weights['model']
                elif isinstance(rad_weights, dict) and 'state_dict' in rad_weights:
                    rad_weights = rad_weights['state_dict']

                # Load the weights into the encoder, ignoring non-matching keys (like 'fc')
                # `strict=False` allows loading only the parts that match our features_only encoder
                missing_keys, unexpected_keys = self.encoder.load_state_dict(rad_weights, strict=False)

                print("RadImageNet Weights Loaded:")
                if missing_keys:
                    print(f"  > Missing keys in model state_dict: {len(missing_keys)}") # Expected if decoder/final layers missing
                if unexpected_keys:
                    print(f"  > Unexpected keys in checkpoint: {len(unexpected_keys)}") # Expected: 'fc.weight', 'fc.bias' etc.
                    # print(f"    Examples: {unexpected_keys[:5]}") # Uncomment to see examples

                if not missing_keys and not unexpected_keys:
                     print("  > Weights loaded successfully with exact match (unexpected for features_only).")
                elif not missing_keys and unexpected_keys:
                     print("  > Weights loaded successfully. Ignored keys likely belong to classifier head.")
                else:
                     # If there are missing keys other than the decoder, it might be an issue
                     print("  > Warning: Some keys might be missing from the encoder itself. Check model compatibility.")


            except Exception as e:
                print(f"Error loading RadImageNet weights from {radimagenet_weights_path}: {e}")
                print("Proceeding with randomly initialized encoder.")
        else:
            if radimagenet_weights_path:
                 print(f"RadImageNet weights file not found at: {radimagenet_weights_path}")
            print("Proceeding with randomly initialized encoder (or timm default if pretrained=True was set).")


        # 3. Define Decoder structure (same as before)
        feature_info = self.encoder.feature_info.get_dicts(keys=['num_chs', 'reduction'])
        chs = [info['num_chs'] for info in feature_info]
        if len(chs) < 5:
            raise ValueError(f"Timm InceptionV3 features_only=True returned only {len(chs)} feature maps. Expected at least 5.")
        # Assuming indices 1, 2, 3, 4 correspond to strides 4, 8, 16, 32
        c0, c1, c2, c3 = chs[1], chs[2], chs[3], chs[4]

        self.dec4 = DecoderBlock(c3,         512)
        self.dec3 = DecoderBlock(512 + c2,   256)
        self.dec2 = DecoderBlock(256 + c1,   128)
        self.dec1 = DecoderBlock(128 + c0,    64)
        self.final_up   = nn.Upsample(scale_factor=2, mode="bilinear", align_corners=False)
        self.final_conv = nn.Conv2d(64, 1, 1)

    def forward(self, x):
        # Forward pass (same as before)
        feats = self.encoder(x);
        if len(feats) < 5: raise ValueError(f"Encoder returned only {len(feats)} feature maps.")
        f0, f1, f2, f3 = feats[1], feats[2], feats[3], feats[4]; d4 = self.dec4(f3); f2r = F.interpolate(f2, size=d4.shape[2:], mode="bilinear", align_corners=False); d3 = self.dec3(torch.cat([d4, f2r], 1)); f1r = F.interpolate(f1, size=d3.shape[2:], mode="bilinear", align_corners=False); d2 = self.dec2(torch.cat([d3, f1r], 1)); f0r = F.interpolate(f0, size=d2.shape[2:], mode="bilinear", align_corners=False); d1 = self.dec1(torch.cat([d2, f0r], 1)); out = self.final_up(d1); out = F.interpolate(out, size=(IMG_SIZE, IMG_SIZE), mode="bilinear", align_corners=False); out = self.final_conv(out); return torch.sigmoid(out)


# ─── Utility: ellipse → HC & Semi-Axes (unchanged) ──────────────────────────
def fit_ellipse_hc_axes(mask_np, px_sz):
    # (Code is unchanged, omitted for brevity)
    binary_mask = (mask_np > 0.5).astype(np.uint8); cnts, _ = cv2.findContours(binary_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    hc_mm, semi_axis_a_mm, semi_axis_b_mm = 0.0, 0.0, 0.0
    if not cnts: return hc_mm, semi_axis_a_mm, semi_axis_b_mm
    cnt = max(cnts, key=cv2.contourArea)
    if len(cnt) < 5: return hc_mm, semi_axis_a_mm, semi_axis_b_mm
    try:
        (_, _), (MA, ma), _ = cv2.fitEllipse(cnt); a_px = MA / 2.0; b_px = ma / 2.0
        if a_px <= 0 or b_px <= 0: return hc_mm, semi_axis_a_mm, semi_axis_b_mm
        h = ((a_px - b_px) ** 2) / ((a_px + b_px) ** 2); circ_px = np.pi * (a_px + b_px) * (1 + (3 * h) / (10 + np.sqrt(max(0, 4 - 3 * h))))
        hc_mm = circ_px * px_sz; semi_axis_a_mm = a_px * px_sz; semi_axis_b_mm = b_px * px_sz
    except cv2.error: return hc_mm, semi_axis_a_mm, semi_axis_b_mm
    return hc_mm, semi_axis_a_mm, semi_axis_b_mm

# --- Dice Loss Implementation (unchanged) ---
class DiceLoss(nn.Module):
    # (Code is unchanged, omitted for brevity)
    def __init__(self, smooth=1e-6): super().__init__(); self.smooth = smooth
    def forward(self, preds, targets): preds = preds.view(-1); targets = targets.view(-1); intersection = (preds * targets).sum(); dice_score = (2. * intersection + self.smooth) / (preds.sum() + targets.sum() + self.smooth); return 1 - dice_score

# --- Segmentation Metrics Calculation (unchanged) ---
def calculate_segmentation_metrics(preds, targets, threshold=0.5, smooth=1e-6):
    # (Code is unchanged, omitted for brevity)
    preds = (preds > threshold).float(); targets = targets.float(); preds_flat = preds.view(preds.shape[0], preds.shape[1], -1); targets_flat = targets.view(targets.shape[0], targets.shape[1], -1); tp = (preds_flat * targets_flat).sum(dim=2); fp = ((1 - targets_flat) * preds_flat).sum(dim=2); fn = (targets_flat * (1 - preds_flat)).sum(dim=2); tn = ((1 - targets_flat) * (1 - preds_flat)).sum(dim=2); dice_score = (2. * tp + smooth) / (2. * tp + fp + fn + smooth); iou_score = (tp + smooth) / (tp + fp + fn + smooth); accuracy = (tp + tn + smooth) / (tp + tn + fp + fn + smooth); precision = (tp + smooth) / (tp + fp + smooth); recall = (tp + smooth) / (tp + fn + smooth); return {"dice": dice_score.mean().item(), "iou": iou_score.mean().item(), "accuracy": accuracy.mean().item(), "precision": precision.mean().item(), "recall": recall.mean().item()}

# ─── Training & Validation (MODIFIED to pass weights path to model) ─────────
def train_and_validate():
    df, px_col, hc_col = load_csv(TRAIN_CSV, want_hc=True)
    all_imgs = sorted(glob.glob(os.path.join(TRAIN_DIR, "**", "*_HC.png"), recursive=True))
    valid_files = [p for p in all_imgs if os.path.basename(p) in df.index]
    print(f"Found {len(valid_files)} images listed in CSV out of {len(all_imgs)} total files found.")
    if not valid_files: raise ValueError(f"No valid training images found matching CSV index in {TRAIN_DIR}.")

    split   = int(0.8*len(valid_files)); train_p = valid_files[:split]; val_p = valid_files[split:]
    print(f"Train samples: {len(train_p)}, Validation samples: {len(val_p)}")

    train_ds = HCDataset(train_p, df, px_col, hc_col, "train", img_size=IMG_SIZE, norm_mean=NORM_MEAN, norm_std=NORM_STD)
    val_ds   = HCDataset(val_p,   df, px_col, hc_col, "val", img_size=IMG_SIZE, norm_mean=NORM_MEAN, norm_std=NORM_STD)

    num_w = 2 if DEVICE == torch.device("cuda") else 0
    dl_train = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=num_w, pin_memory=(num_w>0), collate_fn=collate_fn_skip_none)
    dl_val   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=num_w, pin_memory=(num_w>0), collate_fn=collate_fn_skip_none)

    # --- Instantiate model, passing the weights path ---
    model = UNetInceptionV3(radimagenet_weights_path=RADIMAGENET_INCEPTIONV3_PATH).to(DEVICE)

    optimizer = optim.Adam(model.parameters(), lr=LR)
    bce_loss_fn = nn.BCELoss()
    dice_loss_fn = DiceLoss().to(DEVICE)
    scheduler = lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.1, patience=5, verbose=True)

    best_val_dice = -1.0; best_model_state = None
    model_save_path = "hc_unet_radimagenet_best_dice.pth" # New save name

    history = {'epoch': [], 'train_loss': [], 'val_loss': [], 'val_dice': [], 'val_iou': [], 'val_accuracy': [], 'val_precision': [], 'val_recall': [], 'val_mae': []}
    visualized_batch = False # Debug flag

    for ep in range(1, EPOCHS + 1):
        # --- Training Loop (unchanged) ---
        model.train(); total_train_loss = 0.0; total_train_batches = 0
        train_loop = tqdm(dl_train, desc=f"Epoch {ep}/{EPOCHS} [Train]", leave=False)
        for batch_idx, batch in enumerate(train_loop):
            if batch is None: continue
            imgs, masks, fns, *_ = batch; imgs, masks = imgs.to(DEVICE), masks.to(DEVICE)
            if ep == 1 and not visualized_batch and batch_idx == 0: # Debug save
                print("\nSaving first training batch for visualization..."); debug_dir = "debug_batch_viz"; os.makedirs(debug_dir, exist_ok=True); denormalize = transforms.Normalize(mean=[-m/s for m, s in zip(NORM_MEAN, NORM_STD)], std=[1/s for s in NORM_STD])
                for i in range(min(imgs.size(0), 4)): img_to_save = denormalize(imgs[i].cpu()).permute(1, 2, 0).numpy(); img_to_save = np.clip(img_to_save, 0, 1); mask_to_save = masks[i].cpu().squeeze().numpy(); fname_base = os.path.splitext(fns[i])[0]; img_pil = Image.fromarray((img_to_save[:,:,0] * 255).astype(np.uint8)); mask_pil = Image.fromarray((mask_to_save * 255).astype(np.uint8)); img_pil.save(os.path.join(debug_dir, f"{fname_base}_img_input_ep1.png")); mask_pil.save(os.path.join(debug_dir, f"{fname_base}_mask_input_ep1.png"))
                visualized_batch = True; print(f"...done saving batch images/masks to '{debug_dir}' directory.")
            optimizer.zero_grad(); preds = model(imgs); bce_loss = bce_loss_fn(preds, masks); dice_loss = dice_loss_fn(preds, masks); loss = bce_loss + DICE_WEIGHT * dice_loss
            loss.backward(); optimizer.step(); total_train_loss += loss.item(); total_train_batches += 1
            train_loop.set_postfix(loss=loss.item(), bce=bce_loss.item(), dice=dice_loss.item())
        avg_train_loss = total_train_loss / total_train_batches if total_train_batches > 0 else 0
        current_lr = optimizer.param_groups[0]['lr']; print(f"Epoch {ep}/{EPOCHS} ▶ Train loss: {avg_train_loss:.4f} (LR: {current_lr:.1e})")

        # --- Validation Loop (unchanged) ---
        model.eval(); val_errs = []; batch_metrics_collector = {k: [] for k in ['dice', 'iou', 'accuracy', 'precision', 'recall']}
        total_val_loss = 0.0; total_val_batches = 0
        val_loop = tqdm(dl_val, desc=f"Epoch {ep}/{EPOCHS} [Val]", leave=False)
        with torch.no_grad():
            for batch in val_loop:
                if batch is None: continue
                imgs, masks_gt, fns, pxs, gts = batch; imgs, masks_gt = imgs.to(DEVICE), masks_gt.to(DEVICE)
                pred_masks_tensor = model(imgs); val_bce = bce_loss_fn(pred_masks_tensor, masks_gt); val_dice_loss = dice_loss_fn(pred_masks_tensor, masks_gt)
                val_loss = val_bce + DICE_WEIGHT * val_dice_loss; total_val_loss += val_loss.item(); total_val_batches += 1
                metrics = calculate_segmentation_metrics(pred_masks_tensor, masks_gt)
                for key in batch_metrics_collector: batch_metrics_collector[key].append(metrics[key])
                pred_masks_np = pred_masks_tensor.cpu().numpy()
                for i in range(len(fns)):
                    gt_hc_val = gts[i].item() if isinstance(gts[i], torch.Tensor) else float(gts[i])
                    px_sz_val = pxs[i].item() if isinstance(pxs[i], torch.Tensor) else float(pxs[i])
                    if hc_col is not None and gt_hc_val > 0: pred_hc, _, _ = fit_ellipse_hc_axes(pred_masks_np[i, 0], px_sz_val); val_errs.append(abs(pred_hc - gt_hc_val))
        avg_val_loss = total_val_loss / total_val_batches if total_val_batches > 0 else 0
        avg_metrics = {f"val_{key}": np.mean(values) if values else 0 for key, values in batch_metrics_collector.items()}
        avg_val_mae = np.mean(val_errs) if val_errs and hc_col else -1.0
        print(f"Epoch {ep}/{EPOCHS} ▶ Val Loss: {avg_val_loss:.4f} | Dice: {avg_metrics['val_dice']:.4f}, IoU: {avg_metrics['val_iou']:.4f}, Acc: {avg_metrics['val_accuracy']:.4f}, Prec: {avg_metrics['val_precision']:.4f}, Rec: {avg_metrics['val_recall']:.4f}")
        if avg_val_mae >= 0: print(f"                 ▶ Val MAE: {avg_val_mae:.2f} mm")
        history['epoch'].append(ep); history['train_loss'].append(avg_train_loss); history['val_loss'].append(avg_val_loss)
        for key in avg_metrics: history[key].append(avg_metrics[key])
        history['val_mae'].append(avg_val_mae if avg_val_mae >= 0 else np.nan)
        scheduler.step(avg_metrics['val_dice']); current_dice = avg_metrics['val_dice']
        if current_dice > best_val_dice: best_val_dice = current_dice; best_model_state = copy.deepcopy(model.state_dict()); print(f"   -> New best Val Dice: {best_val_dice:.4f}. State captured.")
        print("-" * 30)

    print(f"\nTraining finished. Best validation Dice: {best_val_dice:.4f}")
    if best_model_state is not None: print(f"Saving best model weights to {model_save_path}"); torch.save(best_model_state, model_save_path); model.load_state_dict(best_model_state)
    else: print("Warning: No best model state was captured. Using weights from the last epoch.")
    return model, val_ds, train_ds, history

# ─── Test‐Set Inference (unchanged) ──────────────────────────────────────────
def inference_on_test(model):
    # (Code is unchanged, omitted for brevity)
    try: df_test, px_col, _ = load_csv(TEST_CSV, want_hc=False)
    except FileNotFoundError as e: print(e); return None
    all_test = sorted(glob.glob(os.path.join(TEST_DIR, "**", "*_HC.png"), recursive=True)); valid_t  = [p for p in all_test if os.path.basename(p) in df_test.index]
    print(f"\nFound {len(valid_t)} test images listed in CSV out of {len(all_test)} total files found.")
    if not valid_t: print("No valid test images found matching CSV. Skipping inference."); return None
    test_ds = HCDataset(valid_t, df_test, px_col, None, "test", img_size=IMG_SIZE, norm_mean=NORM_MEAN, norm_std=NORM_STD); num_w = 2 if DEVICE == torch.device("cuda") else 0
    dl_test = DataLoader(test_ds, batch_size=1, shuffle=False, num_workers=num_w, collate_fn=collate_fn_skip_none)
    model.eval(); rows=[]; test_loop = tqdm(dl_test, desc="Test Inference")
    with torch.no_grad():
        for batch in test_loop:
            if batch is None: continue
            imgs, _, fns, pxs, _ = batch; imgs = imgs.to(DEVICE); pred_mask_tensor = model(imgs); pred_mask_np = pred_mask_tensor.cpu().numpy()[0, 0]; px_sz_val = pxs.item() if isinstance(pxs, torch.Tensor) else float(pxs)
            predicted_hc, pred_a_mm, pred_b_mm = fit_ellipse_hc_axes(pred_mask_np, px_sz_val)
            rows.append((fns[0], predicted_hc, pred_a_mm, pred_b_mm))
    output_df = pd.DataFrame(rows, columns=["filename", "predicted_HC_mm", "semi_axis_a_mm", "semi_axis_b_mm"]); output_csv_path = "test_predictions_with_axes.csv" # Keep same name or change if desired
    output_df.to_csv(output_csv_path, index=False); print(f"✅ Test predictions saved to {output_csv_path}"); return test_ds

# --- Plotting Function (Sample Predictions - unchanged) ---
def plot_predictions(model, train_ds, val_ds, test_ds, num_examples=5, device=DEVICE, norm_mean=None, norm_std=None):
    # (Code is unchanged, omitted for brevity)
    print(f"\nGenerating {num_examples} plot examples for Train, Validation, and Test sets..."); model.eval(); datasets = {"Train": train_ds, "Validation": val_ds, "Test": test_ds}
    num_valid_datasets = sum(1 for ds in datasets.values() if ds is not None and len(ds) > 0);
    if num_valid_datasets == 0: print("No valid datasets with samples found for plotting."); return
    min_len = min((len(ds) for ds in datasets.values() if ds is not None and len(ds) > 0), default=0); num_examples = min(num_examples, min_len)
    if num_examples == 0: print("Not enough samples in datasets to plot."); return
    num_rows = num_valid_datasets * num_examples; fig, axes = plt.subplots(num_rows, 3, figsize=(12, num_rows * 4))
    if num_rows == 1: axes = axes.reshape(1, -1)
    denormalize = transforms.Normalize(mean=[-m/s for m, s in zip(norm_mean, norm_std)], std=[1/s for s in norm_std]); current_row = 0
    with torch.no_grad():
        for set_name, ds in datasets.items():
            if ds is None or len(ds) == 0: continue
            print(f"  Plotting {set_name} examples..."); subset_indices = list(range(num_examples)); subset = torch.utils.data.Subset(ds, subset_indices); dl_subset = DataLoader(subset, batch_size=1, shuffle=False, collate_fn=collate_fn_skip_none)
            for i, batch in enumerate(dl_subset):
                 if batch is None: continue
                 img_tensor, mask_gt, fn, _, _ = batch; img_tensor_dev = img_tensor.to(device); pred_mask = model(img_tensor_dev).cpu(); img_denorm = denormalize(img_tensor.squeeze(0)).permute(1, 2, 0).numpy(); img_denorm = np.clip(img_denorm, 0, 1); img_display = img_denorm[:, :, 0]; gt_np = mask_gt.squeeze().numpy(); pred_np = (pred_mask > 0.5).squeeze().numpy().astype(float); sample_dice = calculate_segmentation_metrics(pred_mask, mask_gt)['dice']
                 ax_img = axes[current_row, 0]; ax_gt = axes[current_row, 1]; ax_pred = axes[current_row, 2]; ax_img.imshow(img_display, cmap='gray'); ax_img.set_title(f"{set_name} {i+1}: Image\n{fn[0]}"); ax_img.axis('off')
                 if set_name in ["Train", "Validation"]: ax_gt.imshow(gt_np, cmap='gray'); ax_gt.set_title(f"Ground Truth")
                 else: ax_gt.text(0.5, 0.5, 'N/A', ha='center', va='center', transform=ax_gt.transAxes); ax_gt.set_title(f"Ground Truth")
                 ax_gt.axis('off'); ax_pred.imshow(pred_np, cmap='gray'); ax_pred.set_title(f"Prediction (Dice={sample_dice:.3f})"); ax_pred.axis('off'); current_row += 1
    plt.tight_layout(pad=1.0, h_pad=1.5); plt.savefig("segmentation_examples_radimagenet.png"); print("✅ Plot examples saved to segmentation_examples_radimagenet.png") # New plot name

# --- Function to plot training curves (unchanged) ---
def plot_training_curves(history):
    # (Code is unchanged, omitted for brevity)
    epochs = history['epoch']; plt.style.use('seaborn-v0_8-darkgrid'); fig, axes = plt.subplots(3, 1, figsize=(10, 15))
    axes[0].plot(epochs, history['train_loss'], label='Training Loss', marker='o', linestyle='--'); axes[0].plot(epochs, history['val_loss'], label='Validation Loss', marker='o', linestyle='-'); axes[0].set_xlabel("Epoch"); axes[0].set_ylabel("Loss (Combined)"); axes[0].set_title("Training & Validation Loss"); axes[0].legend(); axes[0].grid(True)
    axes[1].plot(epochs, history['val_dice'], label='Validation Dice Score', marker='o', linestyle='-', color='green'); axes[1].set_xlabel("Epoch"); axes[1].set_ylabel("Dice Score"); axes[1].set_title("Validation Dice Score"); axes[1].legend(); axes[1].grid(True); axes[1].set_ylim(0, 1)
    valid_mae_epochs = [e for e, m in zip(epochs, history['val_mae']) if not np.isnan(m)]; valid_mae_values = [m for m in history['val_mae'] if not np.isnan(m)]
    if valid_mae_values: axes[2].plot(valid_mae_epochs, valid_mae_values, label='Validation MAE (mm)', marker='o', linestyle='-', color='red'); axes[2].set_xlabel("Epoch"); axes[2].set_ylabel("MAE (mm)"); axes[2].set_title("Validation Head Circumference MAE"); axes[2].legend(); axes[2].grid(True)
    else: axes[2].text(0.5, 0.5, 'MAE not calculated or available.', ha='center', va='center', transform=axes[2].transAxes); axes[2].set_title("Validation Head Circumference MAE")
    plt.tight_layout(pad=2.0); plt.savefig("training_curves_radimagenet.png"); print("✅ Training curves saved to training_curves_radimagenet.png") # New plot name


# ─── Execute ─────────────────────────────────────────────────────────────────
if __name__ == "__main__":
    # (Code is unchanged, omitted for brevity)
    paths_ok = True
    for p in [TRAIN_DIR, TEST_DIR, TRAIN_CSV, TEST_CSV, RADIMAGENET_INCEPTIONV3_PATH]: # Added check for weights file
        if not os.path.exists(p): print(f"ERROR: Path not found: {p}"); paths_ok = False
    if not paths_ok: print("Please fix the paths and try again.")
    else:
        trained_model, validation_ds, training_ds, training_history = train_and_validate()
        if trained_model:
            test_ds = inference_on_test(trained_model)
            plot_predictions(trained_model, training_ds, validation_ds, test_ds, num_examples=NUM_PLOT_EXAMPLES, device=DEVICE, norm_mean=NORM_MEAN, norm_std=NORM_STD)
            plot_training_curves(training_history)

```
